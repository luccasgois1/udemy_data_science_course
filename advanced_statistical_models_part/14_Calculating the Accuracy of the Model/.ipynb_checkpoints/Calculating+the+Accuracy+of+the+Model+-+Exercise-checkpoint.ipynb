# Calculating the Accuracy of the Model

Using the same dataset, expand the model by including all other features into the regression. 

Moreover, calculate the accuracy of the model and create a confusion matrix

## Import the relevant libraries

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sms
sms.set()
```

## Load the data

Load the ‘Bank_data.csv’ dataset.

```python
pathfile = 'Bank_data.csv'
data = pd.read_csv(pathfile,index_col=0)
data
```

```python
data['y'] = data['y'].map({'yes':1,'no':0})
data
```

```python
data.describe()
```

### Declare the dependent and independent variables

Use 'duration' as the independet variable.

```python
x1 = data['duration']
y = data['y']
```

### Simple Logistic Regression

Run the regression and graph the scatter plot.

```python
x = sm.add_constant(x1)
reg_log = sm.Logit(y,x)
results_log = reg_log.fit()
results_log.summary()
```

```python
plt.scatter(x1,y)
plt.ylabel('y')
plt.xlabel('Duration')
plt.show()
```

## Expand the model

We can be omitting many causal factors in our simple logistic model, so we instead switch to a multivariate logistic regression model. Add the ‘interest_rate’, ‘march’, ‘credit’ and ‘previous’ estimators to our model and run the regression again. 

### Declare the independent variable(s)

```python
predictors = ['interest_rate','march','credit','previous','duration']
x2 = data[predictors]
```

```python
xnew = sm.add_constant(x2)
reg_lognew = sm.Logit(y,xnew)
results_lognew = reg_lognew.fit()
results_lognew.summary()
```

### Confusion Matrix

Create the confusion matrix of the model and estimate its accuracy. 

```python
c_table = results_log.pred_table()
accuracy_log = c_table.diagonal().sum()/c_table.sum()
print(accuracy_log)
```

```python
c_tablenew = results_lognew.pred_table()
accuracy_lognew = c_tablenew.diagonal().sum()/c_tablenew.sum()
print(accuracy_lognew)
```

Adding the other varaibles to the model increase the accuracy of the model to 69.3% to 86.29% over 17% higher
